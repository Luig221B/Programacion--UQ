```python
import numpy as np
```

```python

def generador_taylor(funcion, x, terminos, **kwargs):
    # Define una función que acepta valores numéricos
    def funcion_numerica(x_value):
        return eval(funcion, {'x': x_value})
    
    # Calcula la aproximación de la serie de Taylor
    taylor_aproximacion = 0
    for n in range(terminos):
        derivada_n = funcion_numerica(x) / np.math.factorial(n)
        taylor_aproximacion += derivada_n * (x - x) ** n
    
    return taylor_aproximacion

# Solicitar al usuario la función, el valor de x, la cantidad de términos y otros parámetros
funcion = input("Ingrese la función matemática (utilice 'x' como variable independiente): ")
x = float(input("Ingrese el valor de x: "))
terminos = int(input("Ingrese la cantidad de términos: "))

# Llamar a la función generadora de la serie de Taylor
estimacion = generador_taylor(funcion, x, terminos)

# Mostrar la estimación basada en la serie de Taylor
print(f"Estimación basada en la serie de Taylor: {estimacion}")



#se hace con sympy, ver si se puede solucionar el error 
```
