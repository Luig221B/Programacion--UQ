```python
import numpy as np
```

```python

def generador_taylor(funcion, x, terminos, **kwargs):
    # Define una funcin que acepta valores numricos
    def funcion_numerica(x_value):
        return eval(funcion, {'x': x_value})
    
    # Calcula la aproximacin de la serie de Taylor
    taylor_aproximacion = 0
    for n in range(terminos):
        derivada_n = funcion_numerica(x) / np.math.factorial(n)
        taylor_aproximacion += derivada_n * (x - x) ** n
    
    return taylor_aproximacion

# Solicitar al usuario la funcin, el valor de x, la cantidad de trminos y otros parmetros
funcion = input("Ingrese la funcin matemtica (utilice 'x' como variable independiente): ")
x = float(input("Ingrese el valor de x: "))
terminos = int(input("Ingrese la cantidad de trminos: "))

# Llamar a la funcin generadora de la serie de Taylor
estimacion = generador_taylor(funcion, x, terminos)

# Mostrar la estimacin basada en la serie de Taylor
print(f"Estimacin basada en la serie de Taylor: {estimacion}")



#se hace con sympy, ver si se puede solucionar el error 
```
