```python
import pandas as pd
import numpy as  np 
df = pd.read_csv('datos_fisica.csv', sep=';', encoding='ISO-8859-1')
```

```python
def analizar_datos(dataframe, *args,**kwargs):
       for columna in args:
        if columna in dataframe.columns:
            print(f"Procesando la columna '{columna} :")
            operaciones = kwargs.get(columna,[])
            if 'media' in operaciones:
                media = dataframe[columna].mean()
                print(f"Media: {media}")
            if 'mediana' in operaciones: 
                print(f" mediana: {dataframe[columna].mean()}")         
            if 'mediana' in operaciones: 
                print(f" desviacion estandar: {dataframe[columna].std()}") 

        else:
            print(f"la columna '{columna} no existe en el DataFrame")

df['F(N)'] = df['F(N)'].str.replace(',', '.').astype(float) #para solucionar un error de comas en puntos decimales          
analizar_datos(df, 'F(N)', **{'F(N)': ['media', 'mediana', 'desviacion estandar']})  

```
