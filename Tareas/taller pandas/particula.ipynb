```python
class Particula:
    def __init__(self, masa, carga, **kwargs):
        self.carga = carga
        try:
            if masa > 0:
                self.masa = masa
            else:
                raise ValueError("La masa debe ser un nmero positivo.")
        except TypeError:
            raise ValueError("La masa debe ser un nmero positivo.")
        
        self.posicion = kwargs.get('posicion', (0, 0, 0))
        
    def calcular_energia_cinetica(self, velocidad):
        if not isinstance(velocidad, (int, float)):
            raise ValueError("La velocidad debe ser un nmero.")
        
        if velocidad < 0:
            raise ValueError("La velocidad debe ser un nmero positivo.")
        
        energia_cinetica = 0.5 * self.masa * velocidad**2
        return energia_cinetica

# Ejemplo de uso
try:
    particula = Particula(masa=5, carga=2, posicion=(1, 2, 3))
    velocidad = 10  # Velocidad arbitraria
    energia_cinetica = particula.calcular_energia_cinetica(velocidad)
    print(f"Energa Cintica de la Partcula: {energia_cinetica}")
except ValueError as e:
    print(f"Error: {e}")
```
